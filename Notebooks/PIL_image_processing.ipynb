## PIL 
### Let's see how we can use Python Imaging Library for Image preprocessing

```python
from PIL import Image
```

```python
img1 = Image.open('../images/dogs/dog1.jpg')
```

```python
# This will open the image in the full cell
#img1
```

```python
plt.imshow(img1)
```

```python
### Once again the properties
img1.size,img1.format,img1.mode
```

## Image Cropping

```python
left, top, right, bottom = 1250, 500, 4000, 3000
img1_crop = img1.crop((left, top, right, bottom))
plt.imshow(img1_crop)
```

```python
### Copying image
img1_copy = img1.copy()
plt.imshow(img1_copy)
```

## Transpose

```python
img1_transpose1 = img1.transpose(Image.FLIP_LEFT_RIGHT)
img1_transpose2 = img1.transpose(Image.FLIP_TOP_BOTTOM)
img1_transpose3 = img1.transpose(Image.ROTATE_90)
```

```python
fig = plt.figure(figsize=(12,12))
ax1, ax2, ax3 = fig.subplots(3,1)
ax1.imshow(img1_transpose1)
ax2.imshow(img1_transpose2)
ax3.imshow(img1_transpose3)
```

## Resizing and different resample methods

```python
newsize = (1200,900)
img1_resize1 = img1.resize(newsize)
img1_resize1.size
```

```python
img1.resize?
```

```python
plt.imshow(img1_resize1)
plt.show()
```

```python
newsize = (1200,900)
img1_resize2 = img1.resize(newsize, resample=Image.HAMMING)
img1_resize3 = img1.resize(newsize, resample=Image.BICUBIC)
```

```python
fig = plt.figure(figsize=(12,12))
ax1, ax2 = fig.subplots(1,2)
ax1.imshow(img1_resize2)
ax2.imshow(img1_resize3)
```

```python
### There are other resampling methods other than HAMMING and BICUBIC
```

## Image Enhancement

```python
slow_sc = Image.open('../images/L1_1259231848.png')
```

```python
plt.imshow(slow_sc)
```

```python
slow_sc.size
```

```python
slow_sc_crop = slow_sc.crop((100, 60, 675, 540 ))
plt.imshow(slow_sc_crop)
```

### This looks, good we will try image enhancement on this cropped image of a Slow scattering noise transient at aLIGO detector

```python
from PIL import ImageEnhance
slow_sc_con = ImageEnhance.Contrast(slow_sc_crop).enhance(2.5)
```

```python
plt.imshow(slow_sc_con)
```

```python
slow_sc_brt = ImageEnhance.Brightness(slow_sc_crop).enhance(.75)
```

```python
plt.imshow(slow_sc_brt)
```
